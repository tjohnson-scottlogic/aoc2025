```python
file = open("input.txt", 'r')
#file = open("sample.txt", 'r')

for line in file.readlines():
  data = line.split(',')

first_half = ""
second_half = ""
invalid_id_count = 0
invalid_id_sum = 0

for i in range(len(data)):
  print(f"Range {data[i]}")
  pairs = data[i].split('-')
  #print(pairs[0])
  #print(pairs[1])

  # Find IDs in range
  for j in range (int(pairs[0]), int(pairs[1]) + 1):
    id = str(j)
    #print(id)

    # If even number of digits
    if (len(id) % 2 == 0):
      #print("Even digits")
      digits = len(id)
      first_half = id[:int(digits / 2)]
      second_half = id[int(digits / 2):]
      #print(f"first_half is {first_half}")
      #print(f"second_half is {second_half}")

      if (first_half == second_half):
        print("Found invalid ID")
        print(id)
        invalid_id_count += 1
        invalid_id_sum = invalid_id_sum + j
        #first_half = ""
        #second_half = ""

print(f"invalid_id_count is {invalid_id_count}")
print(f"invalid_id_sum is {invalid_id_sum}")
```
