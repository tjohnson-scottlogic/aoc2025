```python
text_file = open("test.txt", "r")
#text_file = open("sample.txt", "r")
#text_file = open("input.txt", "r")
lines = text_file.readlines()
text_file.close()

highest_joltage = 0
highest_joltage_position = 0
second_digit = 0
total_output = 0

# First line
bank = lines[0]

# Look through every digit except for the last one
for i in range (len(bank) - 1):
  if (int(bank[i]) > highest_joltage):
    highest_joltage = int(bank[i])
    highest_joltage_position = i
  print(f"highest_joltage is {highest_joltage}")

# Look for the highest digit following from the first
for i in range (highest_joltage_position + 1,len(bank)):
  if (int(bank[i]) > second_digit):
    second_digit = int(bank[i])
  print(f"second_digit is {second_digit}")

bank_joltage = str(highest_joltage) + str(second_digit)
print(f"bank_joltage is {bank_joltage}")

total_output = total_output + int(bank_joltage)
print(f"total_output is {total_output}")
```
