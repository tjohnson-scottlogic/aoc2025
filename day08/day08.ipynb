```python
from IPython.display import display, HTML
import time

display("This is some output", clear=True)
time.sleep(0.3)
display("This is some other output that will replace it", clear=True)

while True:
    time.sleep(0.3)
    display(HTML("<span style='color: white; background-color: black;'>Now this is flashing green and white</span>"), clear=True)
    time.sleep(0.3)
    display(HTML("<span style='color: black; background-color: green;'>Now this is flashing green and white</span>"), clear=True)
```
