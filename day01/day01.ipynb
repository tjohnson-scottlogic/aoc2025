```python
# Load list of rotations
#import urllib
#data = urllib.request.urlopen("https://raw.githubusercontent.com/tjohnson-scottlogic/iac2025/refs/heads/main/day01/input.txt")

# First line of the text file is L14, but get b'L14\n' from the above, think it's because by default it's looking for a binary file.  Data contains an HTTP response.  Try copying file to Colab space instead.
#for line in data:
  #print(str(line[1]))

  #user_input = input("Stop")

text_file = open("input.txt", "r")
lines = text_file.readlines()
#print(lines)
#print(len(lines))
text_file.close()

#print(lines[0])

current_dial_position = 50
count_of_zeros = 0

for i in range (0,len(lines)):
  #print(lines[i])
  if(lines[i][0] == "L"):
    print("Left")
    print(lines[i][1:-1])
    current_dial_position = current_dial_position - int(lines[i][1:])
    print(f"current_dial_position is {current_dial_position}")
    if (current_dial_position < 0):
      current_dial_position = 100 + current_dial_position
      print(f"current_dial_position is {current_dial_position}")
    if (current_dial_position == 0):
      count_of_zeros = count_of_zeros + 1
      print(f"count_of_zeros = {count_of_zeros}")

  if(lines[i][0] == "R"):
    print("Right")
    print(lines[i][1:-1])
    current_dial_position = current_dial_position + int(lines[i][1:])
    print(f"current_dial_position is {current_dial_position}")
    if (current_dial_position > 99):
      current_dial_position = current_dial_position - 100
      print(f"current_dial_position is {current_dial_position}")
    if (current_dial_position == 0):
      count_of_zeros = count_of_zeros + 1
      print(f"count_of_zeros = {count_of_zeros}")

print(f"Final number for count_of_zeros is {count_of_zeros}")
```
